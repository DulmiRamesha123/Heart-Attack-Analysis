```python
import pandas as pd

data = pd.read_csv('heart.csv')
#print (data)

pd.set_option("display.max.columns", None)
data.head()
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as se


categories = ["sex", "cp", "fbs", "restecg", "exng", "slp", "caa", "thall", "output"]

data_categories = data.loc[:, categories]
for i in categories:
    plt.figure()
    se.countplot(x = i, data = data_categories, hue = "output")
    plt.title(i)
```

```python
pd.crosstab(data.age,data.output).plot(kind="bar",figsize=(15,8))
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter('age', 'trtbps', data=data[data.output==1], marker = 'x', c = 'green')
plt.scatter('age', 'trtbps', data=data[data.output==0], marker = 'o', c = 'orange')
plt.legend(('Yes', 'No'), title='Output')
plt.xlabel('age', fontsize=12)
plt.ylabel('blood pressure', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter('age', 'chol', data=data[data.output==1], marker = 'x', c = 'blue')
plt.scatter('age', 'chol', data=data[data.output==0], marker = '*', c = 'red')
plt.legend(('Yes', 'No'), title='Output')
plt.xlabel('age', fontsize=12)
plt.ylabel('cholestoral', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter('age', 'thalachh', data=data[data.output==1], marker = 'x', c = 'black')
plt.scatter('age', 'thalachh', data=data[data.output==0], marker = 'o', c = 'orange')
plt.legend(('Yes', 'No'), title='Output')
plt.xlabel('age', fontsize=12)
plt.ylabel('Maximum Heart Rate', fontsize=12)
plt.show()
```

```python
data.age.plot(kind = 'hist', bins = 50, figsize = (12,12))
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```
